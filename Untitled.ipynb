## A simple Altair charting script to illustrative where our competitors are making minority investments.

```python
import pandas as pd
import altair as alt
```

```python
data = pd.DataFrame([
    ['Fintech','PwC',6], 
    ['D/A/AI/Cognitive/Automation','PwC', 3],
    ['Cybersecurity/Data Privacy','PwC', 2],
    ['Cybersecurity/Data Privacy', 'EY', 1],
    ['AR/VR', 'PwC', 2],
    ['Enterprise Collaboration', 'PwC', 1],
    ['Enterprise Saas', 'EY', 2],
    ['Enterprise Saas', 'PwC', 4],
    ['Blockchain', 'PwC', 1],
    ['Blockchain', 'EY', 1],
    ['Blockchain', 'Deloitte', 1],
    ['Educational Tech', 'PwC', 1],
    ['Educational Tech', 'EY', 1],
    ['Health Products', 'EY', 1],
    ['Platform/Marketplace', 'PwC', 1],
    ['Robotics/IoT', 'Deloitte', 1]],
    columns=['Focus Area', 'Competitor', 'Number of Investments'])
```

After loading the data above, the following summarizes how many investments are made in total.

```python
display(data.groupby(['Competitor']).sum())
print('The total number of MI Investments is ' + str(data['Number of Investments'].sum()))
```

```python
points = alt.Chart(data, width=400, height=600).mark_bar().encode(
    x = 'Competitor',
    y = 'Focus Area',
    color = alt.Color('Focus Area', legend=None, scale=alt.Scale(scheme='tableau20'))
).configure_axisY(labelPadding=10, titlePadding=10).configure_axisX(labelAngle=45, labelAlign='left', labelFontSize=12)

display(points)
```

```python
base_chart = alt.Chart(data, width=900, height=500).encode(
    x = 'Focus Area',
    y = 'Competitor',
).properties(
    title = 'Minority Investments by Competitor and Area',
)

bars = base_chart.mark_bar().encode(
    color = alt.Color('Focus Area', legend=None, scale=alt.Scale(scheme='tableau20'))
)

data_labels = base_chart.mark_text(
    align='center',
    baseline='middle',
    color='black'
).encode(
    text='Number of Investments'
)

(bars + data_labels).configure_axisY(labelPadding=10, titlePadding=10).configure_axisX(labelAngle=45, labelAlign='left', labelFontSize=12).configure_title(fontSize=20)
```
